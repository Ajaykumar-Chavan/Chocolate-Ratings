```python
sns.histplot(df['rating'],kde=True,bins=20,color='royalblue')
plt.title('Frequency of Ratings')
plt.show()
```

```python
sns.scatterplot(data=df,x='cocoa_percent',y='rating',color='royalblue')
plt.xticks(rotation=90)
plt.title('Relation Between cocoa_percent and rating')
plt.show()
```

```python
df.columns
```

```python
fig = df.groupby('country_of_bean_origin')['rating'].mean()
plt.bar(fig.index,fig.values,color='royalblue')
plt.title('Average Ratings by Countries of Bean origin ')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Average Ratings')
plt.show()
```

```python
sns.countplot(x='company_location',data=df,order=df['company_location'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Count of chocolate by company location')
plt.show()
```

```python
fig = df.groupby('review_date')['rating'].mean()
plt.plot(fig.index,fig.values,color='blue')
plt.title('Average Ratings by Years ')
plt.xlabel('Year')
plt.ylabel('Average Ratings')
plt.show()
```
